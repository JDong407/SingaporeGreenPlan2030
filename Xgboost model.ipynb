```python
df_encoded = pd.get_dummies(df, columns=['Region'])
```

```python
train = df_encoded[(df_encoded['Year'] < 2023)]
test = df_encoded[(df_encoded['Year'] ==2023)]
```

```python
train
```

```python
test
```

```python
X_train = train.drop(["Mean Temperature (°C)"], axis=1)
y_train = train["Mean Temperature (°C)"]

X_test = test.drop(["Mean Temperature (°C)"], axis=1)
y_test = test["Mean Temperature (°C)"]
```

```python
X_test
```

```python
import xgboost as xgb
from sklearn.metrics import mean_squared_error

# Create regression matrices
dtrain_reg = xgb.DMatrix(X_train, y_train)
dtest_reg = xgb.DMatrix(X_test, y_test)

# Define hyperparameters for basic model
params = {"objective": "reg:squarederror"}

n = 100
model = xgb.train(
   params=params,
   dtrain=dtrain_reg,
   num_boost_round=n,
)


y_pred = model.predict(dtest_reg)
# Calculate the mean squared error to evaluate the model
mse = mean_squared_error(y_test, y_pred)
print("Mean Squared Error:", mse)
```

```python
import xgboost as xgb
from sklearn.model_selection import GridSearchCV

# Define the XGBoost parameter grid, including basic model parameters
param_grid = {
    'max_depth': [3, 5, 6, 7],
    'min_child_weight': [1, 3, 5],
    'gamma': [0, 0.1, 0.3, 0.5],
    'subsample': [0.6, 0.7, 0.8, 1],
    'colsample_bytree': [0.6, 0.7, 0.8, 1],
    'learning_rate': [0.1, 0.3, 0.5],
    'n_estimators': [100, 1000],
    
    
}

xgb_regressor = xgb.XGBRegressor(random_state=42)

# Create the GridSearchCV object
grid_search = GridSearchCV(xgb_regressor, param_grid, cv=5, scoring='neg_mean_squared_error')

# Fit the GridSearchCV object to the data
grid_search.fit(X_train, y_train)

# Get the best model and its hyperparameters
best_xgb = grid_search.best_estimator_
best_params = grid_search.best_params_

# Display the best XGBoost model and hyperparameters
print("Best XGBoost Regressor:", best_xgb)
print("Best Hyperparameters:", best_params)
```

```python
y_pred = best_xgb.predict(X_test)

# Evaluate the model on the test data
from sklearn.metrics import mean_squared_error

mse = mean_squared_error(y_test, y_pred)
print("Mean Squared Error on Test Data (XGBoost):", mse)
```
