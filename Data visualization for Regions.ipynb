```python
import seaborn as sns

import pandas as pd

import numpy as np

import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("/Users/melissalian/Desktop/untitled folder/databeforescale.csv")
```

```python
df["Date"] = pd.to_datetime(df.Year.astype(str) + '/' + df.Month.astype(str) + '/01')
```

```python
sns.lineplot(data=df, x="Date", y="Mean Temperature (°C)", hue="Region").set(title='Mean Temperature Across Time for Different Regions')
plt.legend(loc='upper left', bbox_to_anchor=(1, 1))
plt.show()
```

```python
sns.lineplot(data=df, x="Date", y="Daily Rainfall Total (mm)", hue="Region").set(title='Daily Rainfall Total Across Time for Different Regions')
plt.legend(loc='upper left', bbox_to_anchor=(1, 1))
plt.show()
```

```python
sns.lineplot(data=df, x="Date", y="Maximum Temperature (°C)", hue="Region").set(title='Max Temperature Across Time for Different Regions')
plt.legend(loc='upper left', bbox_to_anchor=(1, 1))
plt.show()
```

```python
sns.lineplot(data=df, x="Date", y="Minimum Temperature (°C)", hue="Region").set(title='Min Temperature Across Time for Different Regions')
plt.legend(loc='upper left', bbox_to_anchor=(1, 1))
plt.show()
```

```python
sns.lineplot(data=df, x="Date", y="Mean Wind Speed (km/h)", hue="Region").set(title='Mean Wind Speed Across Time for Different Regions')
plt.legend(loc='upper left', bbox_to_anchor=(1, 1))
plt.show()
```
